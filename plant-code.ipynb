```python
import pandas as pd
```

```python
path = "C:\\Users\\63141\\Desktop\\github_files_v2\\IPM frontend.xlsx"
front_end = pd.read_excel(path)
```

```python
both_path = "C:\\Users\\63141\\Desktop\\github_files_v2\\both.xlsx"
both = pd.read_excel(both_path)
```

```python
# Match the both spreadsheet to match plant codes to IPM front end sheet
```

```python
front_end = front_end.rename(columns= {'Plant':'Plant Name', "Capacity Sub-Type":"Technology"})
both = both.rename(columns= {'Plant Name_x':'Plant Name'})
```

```python
IPM = plant_code_match.dropna(subset=['Plant Code'])
```

```python
pc_path = "C:\\Users\\63141\\Desktop\\github_files_v2\\PC_for_IPM.xlsx"
IPM.to_excel(pc_path, index= False)
```

```python
# now doing a plant code match up from the fuzzy match R1
```

```python
# renaming columns to match each other
# loading in the excel with the technology dictionary for the matching, change path
tech_dict_path = "C:\\Users\\63141\\Desktop\\github files\\tech-lookups.xlsx"
tech_dict_ipm = pd.read_excel(tech_dict_path, sheet_name = "IPM-FE")
```

```python
tech_dict_ipm = tech_dict_ipm.applymap(lambda x: x.lower() if isinstance(x, str) else x)
```

```python
front_end_tech = pd.merge(front_end, tech_dict_ipm, how='left', on='Technology')
```

```python

front_end_tech = front_end_tech.assign(key_column=front_end_tech['Plant Name1'].astype(str) + front_end_tech['Merged Capacity Types'].astype(str))
```

```python
path1 = "C:\\Users\\63141\\Desktop\\github_files_v2\\Matching Different Methods v2.xlsx"
R1 = pd.read_excel(path1, sheet_name= 'R1')
R1 = R1.rename(columns= {'Plant Name_df2':'Plant Name1', "Merged Capacity Types_df2" : "Technology"})
R1 = R1.assign(key_column=R1['Plant Name1'].astype(str) + '_' + R1['Technology'].astype(str))
R1.drop('Plant Name', axis=1, inplace=True)
```

```python
plant_code_matchR1 = front_end_tech.merge(R1[['key_column', 'Plant Code']], 
                                     on=['key_column'], 
                                     how='left')
```

```python
plant_code_matchR1 = plant_code_matchR1.dropna(subset=['Plant Code'])
```

```python
pc_pathR1 = "C:\\Users\\63141\\Desktop\\github_files_v2\\PC_for_IPMR1_v2.xlsx"
plant_code_matchR1.to_excel(pc_pathR1, index= False)
```
